## 2.3 Exercício  003

**Usando a biblioteca numpy**

1. Importe a biblioteca numpy com o alias np
1. Usando os comandos da biblioteca, gere uma variável x contendo 100 valores _float_ linearmente espaçados entre -5 e 5
1. Gere a variável y1 que implemente a função $f_1(x) = x^\frac{n+1}{m+1}$, em que _n_ é o último dígito do seu número de matrícula e m é o penúltimo dígito
1. De forma semelhante, gere a variável y2 que implemente $f_2(x) = (n+1)tanh(x+m)$
1. De forma semelhante, gere a variável y3 que implemente $f_3(x) = e^{-\frac{n+1}{m+1}x}$
1. De forma semelhante, gere a variável y4 que implemente $f_4(x) = e^{\frac{n+1}{m+1}x}$

```python
import numpy as np

x = np.linspace(-5, 5, 100)

n, m = 5, 5

y1 = np.power(x, (n+1)/(m+1))
y2 = (n+1)*np.tanh(x+m)
y3 = np.power(np.e, -(n+1)*x/(m+1))
y4 = np.power(np.e, (n+1)*x/(m+1))
```

## 2.4 Exercício 004

## 2.6 Exercício 006

**Outros gráficos**. Numa mesma figura, usando sub-figuras, imprima os dados anteriores decimados de quatro.
Isto é, a cada quatro pontos do vetor y1 original, por exemplo, o gráfico deve mostrar apenas um ponto. O
vetor a ser plotado terá seu tamanho dividido por quatro ($100 \div 4 = 25$) e, consequentemente, um gráfico
mais "quadrado". O plot deve ser feito usando gráfico com barras verticais ( bar ) para y1 , hastes ( stem )
para y2 , linha tracejada (dashed line) para y3 e area (comando fill_between ) para y4.

```python
mpl.figure(6, dpi=82, figsize=(9,5))
mpl.title('Gráfico $y$ vs $x$')

ax1 = mpl.subplot(411)
mpl.bar(x[::4], y1[::4])
mpl.ylabel('$y_1$')
mpl.setp(ax1.get_xticklabels(), visible=False)

ax2 = mpl.subplot(412, sharex=ax1)
mpl.stem(x[::4], y2[::4])
mpl.ylabel('$y_2$')
mpl.setp(ax2.get_xticklabels(), visible=False)


ax3 = mpl.subplot(413, sharex=ax1)
mpl.plot(x[::4], y3[::4], '--')
mpl.ylabel('$y_3$')
mpl.setp(ax3.get_xticklabels(), visible=False)
mpl.setp(ax3.get_xticklines(), visible=False)

mpl.subplot(414, sharex=ax1)
mpl.fill_between(x[::4], y4[::4])
mpl.xlabel('$x$')
mpl.ylabel('$y_4$')

mpl.show()
```

## 2.8 Exercício 008

**Função**. Faça uma **função** que receba uma matriz bidimensional (como as variáveis Carro1 e Carro2
do exercício anterior) e retorne uma tupla contendo a média e o desvio padrão das amostras contidas nos
dados.

```python
def estatisticas(mat):
    medias = [np.mean(mat[:,0]), np.mean(mat[:,1])]
    desvpads = [np.sqrt(np.var(mat[:,0])), np.sqrt(np.var(mat[:,1]))]

    return (medias[0], medias[1], desvpads[0], desvpads[1])
```

Agora utilize a função desenvolvida nos dados de Carro1 e Carro2 e imprima ( print ) os respectivos
valores. Ao imprimir, adicione um texto explicando os dados exibidos, de qual carro e de qual feature se refere.

```python
stat_carro1 = estatisticas(Carro1)
stat_carro2 = estatisticas(Carro2)

print("Estatísticas referentes ao preço do Carro 1")
print("Média: {} ; Desvio Padrão: {}".format(stat_carro1[0], stat_carro1[2]))
print("\nEstatísticas referentes aos quilômetros rodados do Carro 1")
print("Média: {} ; Desvio Padrão: {}".format(stat_carro1[1], stat_carro1[3]))

print("\n\nEstatísticas referentes ao preço do Carro 2")
print("Média: {} ; Desvio Padrão: {}".format(stat_carro2[0], stat_carro2[2]))
print("\nEstatísticas referentes aos quilômetros rodados do Carro 2")
print("Média: {} ; Desvio Padrão: {}".format(stat_carro2[1], stat_carro2[3]))
```

**Apenas para o preço dos automóveis**, faça num mesmo gráfico (e de cores diferentes) as duas funções
densidade de probabilidade (PDF), baseado na média e desvio padrão encontrados. Considere uma
distribuição normal (gaussiana).

```python
from scipy.stats import norm

dist_carro1 = norm.pdf(Carro1[:, 1], stat_carro1[1], stat_carro1[3])
dist_carro2 = norm.pdf(Carro2[:, 1], stat_carro2[1], stat_carro2[3])

mpl.figure()
mpl.plot(Carro1[:, 1], dist_carro1, 'b.', markersize=3, label="Ford Fusion")
mpl.plot(Carro2[:, 1], dist_carro2, 'r.', markersize=3, label="Fiat Uno")
mpl.xlabel("Preço do carro")
mpl.ylabel("Distribuição normal")
mpl.legend()

mpl.show()
```
